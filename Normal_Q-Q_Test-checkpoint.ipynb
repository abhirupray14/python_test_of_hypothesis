```python
import pandas as mypd
from scipy import stats
import matplotlib.pyplot as myplot
mydata = mypd.read_csv("C:/Users/USER/Data_Sets/PO_Processing.csv")
time = mydata.Processing_Time
stats.probplot(time,plot=myplot)
myplot.show()
result = stats.mstats.normaltest(time)
print(result)
if(result.pvalue >= 0.05):
    print("The given distribution is normal.")
else:
    print("The distribution is not normally distributed.")
```
